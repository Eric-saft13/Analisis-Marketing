1. ACCESO DE DATOS Y PREPARACION PARA ANALISIS

```python
import pandas as pd
import matplotlib.pyplot as plt

visits = pd.read_csv('/datasets/visits_log_us.csv')
orders = pd.read_csv('/datasets/orders_log_us.csv')
costs = pd.read_csv('/datasets/costs_us.csv')
```

```python
visits['Start Ts'] = pd.to_datetime(visits['Start Ts'])
visits['End Ts'] = pd.to_datetime(visits['End Ts'])
visits['Uid'] = pd.to_numeric(visits['Uid'], errors='coerce', downcast='unsigned')
visits['Source Id'] = pd.to_numeric(visits['Source Id'], errors='coerce', downcast='integer')

orders['Buy Ts'] = pd.to_datetime(orders['Buy Ts'])
orders['Revenue'] = pd.to_numeric(orders['Revenue'], errors='coerce')
orders['Uid'] = pd.to_numeric(orders['Uid'], errors='coerce', downcast='unsigned')

costs['dt'] = pd.to_datetime(costs['dt'])
costs['costs'] = pd.to_numeric(costs['costs'], errors='coerce')
costs['source_id'] = pd.to_numeric(costs['source_id'], errors='coerce', downcast='integer')
```

2. INFORMES Y CALCULO DE METRICAS 

```python
# 2.1 Visitas
daily_users = visits.groupby(visits['Start Ts'].dt.date)['Uid'].nunique()
weekly_users = visits.groupby(visits['Start Ts'].dt.to_period('W'))['Uid'].nunique()
monthly_users = visits.groupby(visits['Start Ts'].dt.to_period('M'))['Uid'].nunique()

print('Daily Users')
print(daily_users.head())
print('Weekly Users')
print(weekly_users.head())
print('Monthly Users')
print(monthly_users.head())
```

```python
daily_sessions = visits.groupby(visits['Start Ts'].dt.date).size()
print(daily_sessions.head())
```

```python
visits['session_duration_min'] = (visits['End Ts'] - visits['Start Ts']).dt.total_seconds() / 60
print(visits.head())
```

```python
user_visit_count = visits.groupby('Uid').size().reset_index(name='visit_count')
print(user_visit_count.head())
```

Tambi√©n el c√°lculo de duraci√≥n de sesiones est√° correctamente planteado. El uso de .size() para contar sesiones tambi√©n est√° bien aplicado. Buen enfoque anal√≠tico.

```python
avg_order_value = orders['Revenue'].mean()
print(avg_order_value)
```

```python
ltv = orders.groupby('Uid')['Revenue'].sum().reset_index(name='LTV')
print(ltv.head())
```

```python
top_ltv = ltv.sort_values(by='LTV', ascending=False).head(20)

plt.figure(figsize=(12, 6))
plt.bar(top_ltv['Uid'].astype(str), top_ltv['LTV'])
plt.title('Top 20 Usuarios por LTV')
plt.xlabel('Usuario')
plt.ylabel('LTV ($)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 2.3 Marketing
costs_by_source = costs.groupby('source_id')['costs'].sum().reset_index()
print(costs_by_source)
```

```python
first_visits = visits.sort_values(by='Start Ts').drop_duplicates(subset='Uid')
acquired_users_by_source = first_visits.groupby('Source Id')['Uid'].nunique().reset_index(name='new_users')
print(acquired_users_by_source)
```

```python
cac = costs_by_source.merge(acquired_users_by_source, left_on='source_id', right_on='Source Id')
cac['CAC'] = cac['costs'] / cac['new_users']
print(cac)
```

```python
user_sources = first_visits[['Uid', 'Source Id']]
orders = orders.merge(user_sources, on='Uid', how='left')
revenue_by_source = orders.groupby('Source Id')['Revenue'].sum().reset_index(name='total_revenue')

romi = revenue_by_source.merge(costs_by_source, left_on='Source Id', right_on='source_id')
romi['ROMI'] = (romi['total_revenue'] - romi['costs']) / romi['costs']
print(romi)
```

```python
marketing_summary = cac.merge(romi[['Source Id', 'total_revenue', 'ROMI']], on='Source Id')
print(marketing_summary)
```

```python
marketing_summary_sorted = marketing_summary.sort_values(by='Source Id')

plt.figure(figsize=(8, 4))
plt.bar(marketing_summary_sorted['Source Id'].astype(str), marketing_summary_sorted['CAC'])
plt.title('CAC (Costo de Adquisici√≥n de Clientes) por Fuente')
plt.xlabel('Source Id')
plt.ylabel('CAC ($)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.bar(marketing_summary_sorted['Source Id'].astype(str), marketing_summary_sorted['ROMI'])
plt.title('ROMI (Retorno de la Inversi√≥n en Marketing) por Fuente')
plt.xlabel('Source Id')
plt.ylabel('ROMI')
plt.axhline(0, color='gray', linestyle='--')  # L√≠nea de referencia en 0
plt.tight_layout()
plt.show()
```

<b>√âxito</b> - Excelente interpretaci√≥n del ROMI como retorno sobre inversi√≥n. El gr√°fico con l√≠nea base en 0 ayuda a visualizar claramente qu√© fuentes son rentables y cu√°les no. Buena aplicaci√≥n del pensamiento anal√≠tico para tomar decisiones estrat√©gicas sobre presupuesto.
